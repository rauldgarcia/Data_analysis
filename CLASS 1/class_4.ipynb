Let's obtain the percentage of vote for each candidate in each election

````python
for year in elections['Year'].unique():
    pop_vote = elections[elections['Year'] == year].loc[:, 'Popular vote']
    print([year, pop_vote.values/pop_vote.sum()])
````

Groupby puts together all the elements that have the same indicator

first. return only the first of each group

last return only the last of each group

````python
elections.groupby('Year').first()
````

````python
elections.groupby('Result').first()
````

````python
elections.groupby('Result').last()
````
