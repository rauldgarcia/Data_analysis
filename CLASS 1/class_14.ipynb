Vamos a pasar Damage_Property y Damage_Crops a numeros

```python
storms2013['Damage_Property'].value_counts()
```

```python
storms2013['Damage_Property'] = storms2013['Damage_Property'].str.replace('K', '0').str.replace('M', '0000').str.replace('B', '0000000').str.replace('.', '')
storms2013['Damage_Property'] = storms2013[storms2013['Damage_Property'].notnull()]['Damage_Property'].astype('int64')
```

```python
storms2013.info()
```

```python
storms2013['Damage_Crops'] = storms2013['Damage_Crops'].str.replace('K', '0').str.replace('M', '0000').str.replace('B', '0000000').str.replace('.', '')
storms2013['Damage_Crops'] = storms2013[storms2013['Damage_Crops'].notnull()]['Damage_Crops'].astype('int64')
```

```python
begin_date = pd.to_datetime(storms2013[storms2013['Damage_Property'].isna()]['Begin_Date_Time'])
end_date = pd.to_datetime(storms2013[storms2013['Damage_Property'].isna()]['End_Date_Time'])
(end_date - begin_date).to_frame().value_counts()
```

3913 de los eventos que les faltan datos su duración es 0, posiblemente por eso no tienen datos. 

```python
duracion_faltantes = (end_date - begin_date).to_frame()
duracion_faltantes.sort_values(by=0)
```
