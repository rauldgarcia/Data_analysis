```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('flightsOct.csv')
airports = pd.read_csv('airports.csv')
```

```python
origin = df.groupby('ORIGIN')
```

```python
airports
```

```python
airpotlat = airports[['AIRPORT', 'LATITUDE', 'LONGITUDE']]
airpotlat = airpotlat.rename(columns={'AIRPORT': 'ORIGIN'})
```

Agregamos latitude y longitude de cada origen

```python
dff = df.merge(airpotlat, on='ORIGIN')
```

Eliminamos la columna sobrante

```python
dff = dff.rename(columns={'LATITUDE': 'LATITUDE_ORIGIN', 'LONGITUDE': 'LONGITUDE_ORIGIN'})
dff
```

```python
airpotlat
```

```python
airpotlat = airpotlat.rename(columns={'ORIGIN': 'DEST', 'LATITUDE': 'LATITUDE_DEST', 'LONGITUDE': 'LONGITUDE_DEST'})
airpotlat
```

Agregamos la latitude y longitude del destino

```python
dff = dff.merge(airpotlat, on='DEST')
dff
```

np.arctan2(longitude, latitude)

```python
dff['SIN_AZIMUTH'] = np.sin(np.arctan2((dff['LONGITUDE_DEST'] - dff['LONGITUDE_ORIGIN']), (dff['LATITUDE_DEST'] - dff['LATITUDE_ORIGIN'])))
dff
```

Regresión lineal solo con distancia

```python
bos = origin.get_group('BOS')
```

```python
x = bos.DISTANCE.values.reshape(-1, 1)
y = bos.CRS_ELAPSED_TIME.values.reshape(-1, 1)
basicModel = LinearRegression().fit(x, y)
basicModel.score(x, y)
```

Regresión lineal con distancia y seno de Azimuth

```python
bos_azimuth = dff.groupby('ORIGIN').get_group('BOS')
```

```python
x2 = bos_azimuth[['DISTANCE', 'SIN_AZIMUTH']].values
y2 = bos_azimuth.CRS_ELAPSED_TIME.values.reshape(-1, 1)
model2 = LinearRegression().fit(x2, y2)
model2.score(x2, y2)
```
