```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('mpg')
df = df.dropna()
df.head()
```

```python
x = df.iloc[:,2:6].values
x
```

```python
y = df.iloc[:,0].values.reshape(-1,1)
y.shape
```

```python
n, p =x.shape
```

```python
v1 = np.ones(n)
i = np.identity(n)

h = v1 * np.linalg.inv(v1.T.dot(v1).reshape(-1, 1)) * v1.T
h.shape
```

```python
h.dot(y)
```

```python
y.mean()
```

Centered vector

$y - <y>$

```python
(i - h).dot(y)
```

```python
y - y.mean()
```

Variance

```python
y.T.dot((i - h).dot(y))/(n-1)
```

```python
y.var()
```

Some computional methods use $n$ instead of $n-1$ in the denominator

```python
xc = (i - h).dot(x)
xc
```

```python
x[:,0] - x[:, 0].mean()
```

```python
x.T.dot((i - h))
```

```python
x[0,:] - x[0,:].mean()
```

```python
xcov = x.T.dot((i - h).dot(x)) / (n-1)
xcov
```

```python
np.cov(x.T)
```

If we use np.cov(x) we obtain the covariance between rows

$\beta=(x^T x)^{-1}x^T y$

1. $y_{i} - (\beta x)_{i}$

```python
B = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y).reshape(1, -1)
B
```

```python
y[0] - (B.dot(x[0]))
```

```python
error = y - (B.dot(x.T).reshape(-1, 1))
error
```

```python
sns.scatterplot(error)
```

```python
sns.histplot(error)
```

2. $(y - <y>)_{i} - (\beta x)_{i}$

```python
y_centered = (i - h).dot(y)
y_centered
```

```python
B_y_centered = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y_centered).reshape(1, -1)
B_y_centered
```

```python
error_y_centered = y_centered - (B_y_centered.dot(x.T).reshape(-1, 1))
error_y_centered
```

```python
sns.scatterplot(error_y_centered)
```

```python
sns.histplot(error_y_centered)
```

3. $y_{i} - (\beta (x-<x>))_{i}$

```python
x_centered = (i - h).dot(x)
x_centered
```

```python
B_x_centered = np.linalg.inv(x_centered.T.dot(x_centered)).dot(x_centered.T).dot(y).reshape(1, -1)
B_x_centered
```

```python
error_x_centered = y - (B_x_centered.dot(x_centered.T).reshape(-1, 1))
error_x_centered
```

```python
sns.scatterplot(error_x_centered)
```

```python
sns.histplot(error_x_centered)
```

4. $(y - <y>)_{i} - (\beta (x - <x>))_{i}$

```python
B_both_centered = np.linalg.inv(x_centered.T.dot(x_centered)).dot(x_centered.T).dot(y_centered).reshape(1, -1)
B_both_centered
```

```python
error_both_centered = y_centered - (B_both_centered.dot(x_centered.T).reshape(-1, 1))
error_both_centered
```

```python
sns.scatterplot(error_both_centered)
```

```python
sns.histplot(error_both_centered)
```
