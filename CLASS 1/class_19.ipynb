### Normal distribution
#### November 15th

It's probability density function, pdf, is given mathematically by 

$\rho(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp \left (-\frac{1}{2}\frac{(x-\mu)^2}{\sigma ^2} \right)$

for any positive real number $\sigma, \sigma>0$ and any real number $\mu$.

* $\mu$ is called the location parameter of the distribution
* $\sigma$ the scale parameter

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x_pdf = np.linspace(-5, 5, 250)
y_pdf = stats.norm.pdf(x_pdf, loc=0, scale=1)
n0 = plt.plot(x_pdf, y_pdf)
```

Evaluate the value of the function at $x = \sigma$

```python
x_pdf = np.linspace(-4, 4, 250)
mu = 0.
sig = 1
plt.axvline(mu)
plt.plot(x_pdf, stats.norm.pdf(x_pdf, loc=mu, scale=sig), c='r')
fs = stats.norm.pdf(sig, loc=mu, scale=sig)
plt.axvline(sig, c='k')
plt.axhline(fs, c='k')

plt.axvline(2*sig, c='k')
plt.axhline(stats.norm.pdf(2*sig, loc=mu, scale=sig), c='k')

plt.axvline(3*sig, c='k')
plt.axhline(stats.norm.pdf(3*sig, loc=mu, scale=sig), c='k')
```
