```python
import numpy as np
import pandas as pd
import plotly.express as px
import os
```

```python
eviMean = pd.read_csv('EVImean.csv')
eviMean
```

Change names of a dataframe with rename and a dictionary

```python
import copy
old_index = copy.deepcopy(eviMean.columns.to_list())
old_index
```

```python
test = eviMean.columns[:8].union(eviMean.columns[8:].str.split('X').str[-1], sort=False).to_list()
test
```

```python
dictionary = dict(zip(old_index, test))
dictionary
```

```python
eviMean.rename(columns=dictionary)
```

```python
eviMean.columns[8:].str.split('X').str[-1].str.slice(stop=4).to_frame()
```

```python
year = eviMean.columns[8:].to_frame(name='Year').Year.str.split('X').str[-1].str[:4].to_frame()
year
```

```python
day = eviMean.columns[8:].to_frame(name='Day').Day.str.split('X').str[-1].str[4:].to_frame()
day
```

Join function join two dataframes based on the index

```python
year_day = year.join(day)
year_day
```

If one on the dataframe doesn't have some index, it's puts Nan in that celd

```python
eviMean.transpose().join(year_day)
```

Change names of a dataframe with a list

```python
eviMean = eviMean.set_axis(eviMean.columns[:8].union(eviMean.columns[8:].str.split('X').str[-1], sort=False), axis=1)
eviMean
```

```python
regions = eviMean.groupby(by='DESECON1')
```

```python
regions.groups.keys()
```

```python
reg1 = regions.get_group('Sierras Templadas')
reg1.count()
```

```python
reg1.columns.str[:4]
```

```python
reg1_by_year = reg1.groupby(lambda col: col[:4], axis=1)
```

```python
reg1_by_year.groups.keys()
```

```python
reg1_by_year.get_group('2000')
```

```python
reg1_by_year.get_group('lati')
```

```python
reg1_by_year.get_group('2000').sample(50).transpose().plot(legend=False)
```
